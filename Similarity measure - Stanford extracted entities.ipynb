```python
# Imports

import sys, operator
import pandas as pd
import csv
```

```python
# Read in data - the Stanford extracted entities

with open('four_entity_process_stanford.txt', 'r') as f:
    reader = csv.reader(f)
    file_in = list(reader)
ents_df = pd.DataFrame(file_in)
del ents_df[4]
del ents_df[3]
ents_df
```

```python
ents_df.columns = ['Entity', 'Entity_type', 'Frequency']
ents_df['Entity'] = ents_df['Entity'].str.slice(2, -1) 
ents_df['Entity_type'] = ents_df['Entity_type'].str.slice(2, -2) 
ents_df
```

```python
# Read in the known entities - MP names and Constituencies

# Read in the lookup
import csv
with open('..\\mps.csv', 'r') as f:
    reader = csv.reader(f)
    mp_list = list(reader)

    
    
# Split imported lookup into MP name and Consituency

mp_names = []
mp_consts = []

for row in mp_list[1:]:
#    mp_names.append(list(itertools.chain.from_iterable(row[1:3])))
    temp = [row[1] + ' ' + row[2], 'MP']
    mp_names.append(temp)
    temp2 = [row[4], 'Constituency']
    mp_consts.append(temp2)
```

```python
# Have a look at the output to check

mp_consts
```

```python
# Convert Stanford extracted entities to list

ents_list = ents_df['Entity'].tolist()
ent_type_list = ents_df['Entity_type'].tolist()
```

```python
# Convert reference data to lists - values separate to types

mp_names_list=[]
mp_type_list=[]
consts_names_list=[]
consts_type_list=[]

for row in mp_names:
    mp_names_list.append(row[0])
    mp_type_list.append(row[1])
    
for row in mp_consts:
    consts_names_list.append(row[0])
    consts_type_list.append(row[1])
```

```python
# Print the outputs to check

consts_type_list
```

```python
# Concatenate full lists together

ents_list_full = ents_list + mp_names_list + consts_names_list
ents_type_full = ent_type_list + mp_type_list + consts_type_list
ents_type_full
```

```python
# Functions

def tokenize(s, glen):
    g2 = set()
    if len(s) < glen:
        g2.add(s)
        g2.add(s +' ')
    else:
        for i in range(len(s)-(glen-1)):
            g2.add(s[i:i+glen])
    return g2

def dice_grams(g1, g2):
    return (2.0*len(g1 & g2)) / (len(g1)+len(g2))

def dice(n, s1, s2):
    return dice_grams(tokenize(s1, n), tokenize(s2, n))

def term_similarity(sample, labels):
    GRAM_LEN = 4
    scores = []
    for i in range(0,len(sample)):
        for j in range(i+1, len(sample)):
            s1 = sample[i]            
            s2 = sample[j]
            l1 = labels[i]            
            l2 = labels[j]
            score = dice(GRAM_LEN, s1, s2)
#            print(s1, ":", s2, " = ", score)
            scores.append([s1, l1, s2, l2, score])
    return pd.DataFrame(scores, columns=["Entity1", "Label1", "Entity2", "Label2", "Score"])
```

```python
# Run similarity measure on MPs names

sim_scores = term_similarity(ents_list_full, ents_type_full)
sim_scores
```

```python
# Check Ian Murray as an example

sim_scores.loc[sim_scores['Entity1']=='Ian Murray']
```

```python
# Check Trump as an example

sim_scores.loc[sim_scores['Entity1']=='Trump']
```

```python
# Find top matching Entity2, for each Entity1
# Remove zero rows

hi_score_df = sim_scores.loc[sim_scores['Score']>0].sort_values('Score', ascending=False).drop_duplicates(['Entity1'])

hi_score_df

# sim_scores.loc[sim_scores['Score'] > 0.5].sort_values('Score', ascending=False)
```

```python
# Merge hi score back onto original file

match_df = pd.merge(ents_df, hi_score_df, left_on='Entity', right_on='Entity1', how='left')
match_df
```

```python
# Output file

match_df.to_csv('fuzzy_match_hansard_stanford_ner_170322.csv')
```
